# Um investimento

Esta etapa custará 3 mil reais.

A pessoa decide que pagará á vista, mesmo sabendo que pode parcelar.
Pois tem os 6 mil na reserva.



Pagando este valor á vista estaria consumindo uma boa parte da reserva.


O que acontece com muitos.

# Então comprando a vista:


A pessoa:

* tem os 6 mil
* Paga **à vista** os 3 mil nesta compra
* Fez esta compra no fim de Novembro

* E consideramos que guarda 200 reais por mês.





```python
import matplotlib.pyplot as plt

# plotando compra á vista
meses = ['Dez','Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set']
reserva = 6000
compra = 3000
vezes = 10
taxa_rendimento = .01
aporte = 200
saldo = []

reserva = reserva - compra

print(f'Saldo Inicial: R${reserva}\n\nRENTABILIDADE \n\nPróximos meses: \n')

while vezes >0:
  reserva
  saldo.append(reserva)
  reserva += aporte
  reserva += taxa_rendimento * reserva
  vezes -= 1


plt.bar(meses, saldo)

ax = plt.gca()
ax.spines[['top', 'right']].set_visible(False)
plt.ylim(0, 7000)

for i, v in enumerate(saldo):
  plt.text(i, v+30, f'{v:.0f}', ha='center', size= 8)

plt.show()
```

Após a compra, a reserva retornaria ao valor de R$ 3 mil.

E nos próximos 10 meses atingiria aproximadamente R$ 5173.
Isto , considerando:

* uma rentabilidade de 1% ao mês.
* e a economia de 200 reais por mês.

Vendo desta forma não parece ruim.

Logo há uma motivação em fazer.




#Não estamos habituados a olhar para  a motivação.

Economizar não é simplesmente guardar dinheiro.
Há uma função para este dinheiro - além do obvio que é pagar as compras.




**FUNÇÃO DO DINHEIRO**

Te dar mais dinheiro.

VOU TENTAR TE MOSTRAR!

```python
# reserva mês a mês se pagar parcelado, sem aportes
import datetime
from datetime import timedelta

dt = datetime.date(2023,12,1)
next = timedelta(days=30)
reserva = 6000
parcela = compra/10
vezes = 10
soma= 0
saldo =[]
renda = []

print(f'Sem contar com as reservas que vai fazer!\n\nE sem contar os juros. \
 \n\nSALDO INICIAL: R${reserva} \n\nSaldo nos meses seguintes: \n')

while vezes >0:
  rendimento = reserva * taxa_rendimento
  reserva += rendimento
  print(f'R$ {rendimento:.02f}')
  reserva -= parcela
  saldo.append(reserva)
  vezes -= 1
  soma += rendimento
  final = soma
  renda.append(soma)

print(f'\n\nRendimento Total nos próximos 10 meses: {final:.02f} \n\n')

plt.bar(meses, saldo, color='orange')
plt.ylim(0, 7000)
ax = plt.gca()
ax.spines[['top', 'right']].set_visible(False)

plt.show()
```

Isto seria a pessoa tirando as parcelas da  conta no mês de pagamento.

Pode ser pouco contra intuitivo... ver caindo.

## Vou mostrar agora o mais legal.

### Parcelando e fazendo continuando com as economias.

```python
reserva = 6000
parcela = compra/10
vezes = 10
saldo =[]
renda = []

while vezes >0:
  rendimento = reserva * taxa_rendimento
  reserva += rendimento + aporte
  print(f'R$ {reserva:.02f}')
  reserva -= parcela
  saldo.append(reserva)
  vezes -= 1


plt.bar(meses, saldo, color='orange')
plt.ylim(0, 7000)
ax = plt.gca()
ax.spines[['top', 'right']].set_visible(False)

plt.show()
```

## "Legal, está baixando menos... mas continua caindo!"

Reparem:

#Rentabilidades
---
##Vejamos primeiro a rentabilidade do caso em que a compra foi feita à vista.

Reserva com os R$ 3 mil.


```python
reserva = 6000
taxa_rendimento = .01
soma= 0
renda_mensal = []
lucro = []
mes= 0
reserva -= compra

print(f'Saldo Inicial: R${reserva}\n\nRENTABILIDADE \n\nPróximos meses: \n')

for r in range(10):
      rendimento = reserva * taxa_rendimento
      reserva = reserva + rendimento
      mes += 1
      print(f'{mes}º: R${rendimento:.02f}')
      soma += rendimento
      final = soma
      renda_mensal.append(soma)
      lucro.append(final)

print(f'\n\nRendimento Total nos próximos 10 meses: {final:.02f} \n\n')

plt.bar(meses, renda_mensal)
plt.ylim(0, 600)

for i, r in enumerate(renda_mensal):
  plt.text(i, r+8, f'{r:.0f}', ha= 'center', size=8)


plt.ylim
ax = plt.gca()
lados = ['top', 'right']
for s in lados:
  ax.spines[s].set_visible(False)

plt.annotate('Evolução aproximada do valor total da rentabilidade.\nNo período de 10 meses.',
             xy=(.11,500))


plt.show()

```

## É legal ver a rentabilidade não é verdade?

De que em 10 meses você teria um pouco mais 300 reais sem qualquer esforço.

Apenas por deixar o conjunto de suas economias paradinho lá na conta.

##Agora vejamos a evolução da rentabilidade com:

* a compra parcelada
* mais aquela economia de 200 reais por mês.


```python
meses = ['Dez','Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set']
reserva = 6000
taxa_rendimento = .01
aporte = 200
compra = 3000
vezes = 10
parcela = compra/10
soma = 0
saldo =[]
renda = []
dt = datetime.date(2023,12,1)
next = timedelta(days=30)
mes = 0

print('RENTABILIDADES  \n\n')

while vezes >0:
  rendimento = reserva * taxa_rendimento
  reserva += rendimento + aporte
  dt += next
  mes += 1
  print(f'Mês da {mes}a parcela: R$ {rendimento:.02f}')
  saldo.append(reserva)
  soma += rendimento
  renda.append(soma)
  reserva -= parcela

  vezes -= 1

plt.bar(meses, renda, color='orange')

ax = plt.gca()
ax.spines[['top', 'right']].set_visible(False)
# plt.yticks([])

for i, r in enumerate(renda):
  plt.text(i, r+8, f'{r:.0f}', ha= 'center', size= 8)
```

#Significativa, não? A diferença.

* Pagando a compra em parcelas
* Mais a rentabilidade da reserva
* Mais as economias mês a mês

A rentabilidade não só é maior.

Com a compra parcelada a reserva rende quase o DOBRO...
