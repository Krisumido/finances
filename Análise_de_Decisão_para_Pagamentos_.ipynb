#Crédito ou À vista?

Antes de falar sobre as formas de pagamento vamos relembrar um conceito importante quanto às finanças.




###Reserva de Emergência

RESERRVA.jpg

É um conceito bastante falado  no nosso país. Ainda mais pelo histórico de inadimplência (dívidas) da população.

Na grande maioria a instrução é para construir uma Reserva com o valor de no mínimo 6 meses do seu custo de vida.

Para que em uma eventualidade, como o desemprego, você possa se manter por este período (de 6 meses).

##Pensemos no seguinte caso:


Uma pessoa começa a economizar seu dinheiro há um tempo porque pensava na importância de ter um pouco guardado.

No mais, pensando como um suporte para possíveis imprevistos.

Desde então, tem 6 mil reais em economias, que veio sendo acumulado a um pouco mais de um ano.

O que, diga-se de passagem, claramente não atingiu um valor apropriado para a Reserva.

Então em um determinado plano da vida chega uma etapa de maior custo.




# Um investimento
paynow.png

Esta etapa custará 3 mil reais.

A pessoa decide que pagará á vista, mesmo sabendo que pode parcelar.
Pois tem os 6 mil na reserva.



Pagando este valor á vista estaria consumindo uma boa parte da reserva.


O que acontece com muitos.

Pois a motivação que a maioria tem é: "*Se tenho o dinheiro, logo, vou comprar. Vou guardar para quê?!*"





# Então comprando a vista:


A pessoa:

* tem os 6 mil
* Paga **à vista** os 3 mil nesta compra
* Fez esta compra no fim de Novembro

* E consideramos que guarda 200 reais por mês.





nplt1.png

Após a compra, a reserva retornaria ao valor de R$ 3 mil.

E nos próximos 10 meses atingiria aproximadamente R$ 5173.
Isto , considerando:

* uma rentabilidade de 1% ao mês.
* e a economia de 200 reais por mês.

Vendo desta forma não parece ruim.

Logo há uma motivação em fazer.
